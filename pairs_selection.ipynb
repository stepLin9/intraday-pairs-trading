```python
import yfinance as yf
import pandas as pd
import time

tickers = [
"AAPL","MSFT","AMZN","GOOG","META","NVDA","AMD","INTC",
"JPM","BAC","C","GS","MS",
"KO","PEP","PG","CL","WMT","TGT","COST",
"XOM","CVX","COP","SLB",
"V","MA","AXP",
"DIS","NFLX","CMCSA",
"DAL","UAL","LUV",
"HD","LOW",
"ADBE","CRM","ORCL",
"TXN","QCOM","AVGO",
"PFE","JNJ","MRK",
"T","VZ",
"CAT","BA"
]

data = pd.DataFrame()

for ticker in tickers:

    print("Downloading:", ticker)

    try:
        temp = yf.download(
            ticker,
            start="2022-01-01",
            end="2024-01-01",
            progress=False
        )["Adj Close"]

        data[ticker] = temp

        time.sleep(3)

    except Exception as e:
        print("Failed:", ticker)

data = data.dropna(axis=1)
```
